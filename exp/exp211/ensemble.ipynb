```python
import pandas as pd
```

```python
EXP_NAME = "exp211"
BASE = 4
WEIGHTS = [1.00,1.00,1.00]
```

```python
sub1 = pd.read_csv(f"../../output/exp206_for_ensamble.csv")
sub2 = pd.read_csv(f"../../output/exp208_for_ensamble.csv")
sub3 = pd.read_csv(f"../../output/exp209_for_ensamble.csv")
```

```python
sub1['predict'] = sub1.apply(lambda row: ' '.join(row.astype(str)), axis=1)
sub2['predict'] = sub2.apply(lambda row: ' '.join(row.astype(str)), axis=1)
sub3['predict'] = sub3.apply(lambda row: ' '.join(row.astype(str)), axis=1)
```

```python
def cust_blend(dt, W = [1,1,1], base=3):
    REC = []
    REC.append(dt['predict_0'].split())
    REC.append(dt['predict_1'].split())
    REC.append(dt['predict_2'].split())
    
    res = {}
    for M in range(len(REC)):
        for n, v in enumerate(REC[M]):
            if v in res:
                res[v] += (W[M]/(n+base))
            else:
                res[v] = (W[M]/(n+base))
    
    # Sort dictionary by item weights
    res = list(dict(sorted(res.items(), key=lambda item: -item[1])).keys())
    
    # Return the top 10 itens only
    return ' '.join(res[:10])
```

```python
sub = sub1[['predict']]
sub['predict1'] = sub2['predict']
sub['predict2'] = sub3['predict']
sub.columns = [f'predict_{i}' for i in range(len(sub.columns))]
```

```python
sub
```

```python
sub['predict'] = sub.apply(cust_blend, W = WEIGHTS, base=BASE, axis=1)
```

```python
ensemble_sub = sub['predict'].str.split(' ',expand=True)
```

```python
ensemble_sub.columns = [f'predict_{i}' for i in range(10)]
```

```python
ensemble_sub
```

```python
ensemble_sub.to_csv(f"../../output/{EXP_NAME}.csv",index=False)
```
